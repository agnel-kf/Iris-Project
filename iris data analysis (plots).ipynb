```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
iris = pd.read_csv('iris.csv')
```

```python
attributes = ["sepal_length", "sepal_width", "petal_length", "petal_width", "species"]
iris.columns = attributes
```

```python
iris.species.value_counts()
```

# Task : To classify a flower using the given features

```python
species = ['setosa','virginica','versicolor']
for each in range(3):
    x = iris[iris['species'] == species[each]]
    plt.scatter(x['petal_length'],x['petal_width'],label = species[each])
    
plt.legend()
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
```

## Pair Plots 
### We create pair plots to compare the various possiblities with the features

```python
plt.close()
sns.set_style('darkgrid');
sns.pairplot(iris,hue = 'species',size = 2);
plt.show()
```

* petal length and petal width are the the best pair features to be used.
* we can find lines dividing the species
* can use if statements to create a basic learning model

* Has a major disadvavantage i.e. dimensions. 
* In other words if the features exceeded then it is not feasible

## Univariate Analysis
* In this we basically focus on a single feature

```python
sns.FacetGrid(iris,hue = 'species',size = 5) \
    .map(sns.distplot,'petal_length') \
    .add_legend()
plt.show()
```

```python
sns.FacetGrid(iris,hue = 'species',size = 5) \
    .map(sns.distplot,'petal_width') \
    .add_legend()
plt.show()
```

```python
sns.FacetGrid(iris,hue = 'species',size = 5) \
    .map(sns.distplot,'sepal_length') \
    .add_legend()
plt.show()
```

```python
sns.FacetGrid(iris,hue = 'species',size = 5) \
    .map(sns.distplot,'sepal_width') \
    .add_legend()
plt.show()
```

* We can observe that petal length is slightly better than petal width which is significantly better than sepal length and width.
## Box Plots

```python
sns.boxplot(x = 'species',y = 'petal_length',data = iris,width = 1)
```

* From this we can interpret that approximately 25 percent values of versicolor and virginica coincide 
* So there can be an error of 25 percent
* This can only be inferred using box plots
## Violin Plots


* A violin plot combines the benefits of the previous two plots and simplifies them
* Denser regions of the data are fatter, and sparser ones thinner in a violin plot

```python
sns.violinplot(x = 'species',y = 'petal_length',data = iris,gridsize=800,width = 1)
```

* Again we see that most of the values of setosa are accumulated whereas in case of the versico they are distributed
## Multivariate Probability Density

```python
iris_setosa = iris[iris['species'] == 'setosa']
iris_versicolor = iris[iris['species'] == 'versicolor']
iris_virginica = iris[iris['species'] == 'virginica']



sns.jointplot(x = 'petal_length',y = 'petal_width',data = iris_setosa,kind = 'kde')
plt.show()
```

## Mean,Variance and Std - dev

```python

print('Mean')
print(np.mean(iris_setosa['petal_length']))
print(np.mean(iris_versicolor['petal_length']))
print(np.mean(iris_virginica['petal_length']))

print('\nStd-dev')
print(np.std(iris_setosa['petal_length']))
print(np.std(iris_versicolor['petal_length']))
print(np.std(iris_virginica['petal_length']))
```

## Median,Percentile and Interquartile

```python
print('Median')
print(np.median(iris_setosa['petal_length']))
print(np.median(iris_versicolor['petal_length']))
print(np.median(iris_virginica['petal_length']))

print('\n Percentile')
print(np.percentile((iris_setosa['petal_length']),np.arange(0,100,25))) ### get me all the nos. from 0 to 100 with 25 step
print(np.percentile((iris_versicolor['petal_length']),np.arange(0,100,25)))
print(np.percentile((iris_virginica['petal_length']),np.arange(0,100,25)))

```
